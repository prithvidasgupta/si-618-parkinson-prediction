```python
import pandas as pd

proteins = pd.read_csv('./train_proteins.csv')
```

```python
proteins.head()
```

```python
npx_features = pd.pivot(proteins, index='visit_id', values='NPX', columns='UniProt')
```

```python
npx_features = npx_features.fillna('0')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

npx_features.corr()
```

```python
sns.heatmap(npx_features.corr())
```

```python
clinical_data = pd.read_csv('./train_clinical_data.csv')
```

```python
clinical_data = clinical_data.set_index('visit_id')
```

```python
clinical_data.head()
```

```python
dataset = pd.merge(clinical_data[['updrs_2']], npx_features, how='inner', left_on=['visit_id'], right_on=['visit_id'])
dataset.head()
```

```python
Y = dataset[[dataset.columns[0]]].fillna(0)
X = dataset[dataset.columns[1:]]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_validate

X_scaled = StandardScaler().fit_transform(X.values)
lm =LinearRegression(n_jobs=-1)
cvr = cross_validate(lm, X_scaled, Y.values, cv=5, scoring=('r2'))
```

```python
cvr
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_jobs=-1)

cvr = cross_validate(rf, X_scaled, Y.values, cv=5, scoring=('r2'))
```

```python
cvr
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor

gb = HistGradientBoostingRegressor()

cvr = cross_validate(gb, X_scaled, Y.values, cv=5, scoring=('r2'))
```
